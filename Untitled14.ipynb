Question 1 : What is Dimensionality Reduction? Why is it important in machine
learning?

Answer:
Dimensionality Reduction is the process of reducing the number of input variables (features) in a dataset while preserving as much relevant information as possible.
- It is important because:
- Reduces computational cost and training time.
- Helps avoid the “curse of dimensionality.”
- Improves visualization of high-dimensional data.
- Removes noise and redundant features, leading to better model performance.


Question 2: Name and briefly describe three common dimensionality reduction
techniques..


Answer:
- Principal Component Analysis (PCA): Transforms correlated features into a smaller set of uncorrelated variables called principal components.
- Linear Discriminant Analysis (LDA): Projects data onto a lower-dimensional space while maximizing class separability.
- t-Distributed Stochastic Neighbor Embedding (t-SNE): Non-linear technique that preserves local structure, often used for visualizing high-dimensional data in 2D/3D.



Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Answer:
Clustering is the process of grouping similar data points together without predefined labels.
- Popular algorithms:
- K-Means Clustering – partitions data into k clusters based on distance to centroids.
- Hierarchical Clustering – builds a tree-like structure of nested clusters.
- DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – groups points based on density, identifying clusters of arbitrary shape.


Question 4: Explain the concept of anomaly detection and its significance.

Answer:
Anomaly Detection identifies data points that deviate significantly from the majority.
- Significance:
- Detects fraud in financial transactions.
- Identifies faults in industrial systems.
- Improves cybersecurity by spotting unusual network activity.
- Ensures data quality by flagging outliers.


Question 5: List and briefly describe three types of anomaly detection techniques.

Answer:
- Statistical Methods: Use probability distributions to detect rare events (e.g., Z-score, Gaussian models).
- Machine Learning Methods: Algorithms like Isolation Forest or Autoencoders learn normal patterns and flag deviations.
- Distance/Clustering-Based Methods: Outliers are detected based on distance from cluster centers or density (e.g., DBSCAN).


Question 6: What is time series analysis? Mention two key components of time series data.

Answer:
Time Series Analysis involves studying data points collected over time to identify patterns, trends, and dependencies.
- Key components:
- Trend: Long-term increase or decrease in data.
- Seasonality: Regular, repeating patterns at fixed intervals (e.g., monthly sales peaks).

Question 7: Describe the difference between seasonality and cyclic behavior in time series.

Answer:
- Seasonality: Occurs at fixed, known intervals (e.g., quarterly sales, daily temperature cycles).
- Cyclic Behavior: Irregular fluctuations influenced by external factors like economy or business cycles, with no fixed period.


Question 8: Write Python code to perform K-means clustering on a sample dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

X = np.array([[1,2],[1,4],[1,0],
              [4,2],[4,4],[4,0]])
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

print("Cluster Centers:\n", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            color='black', marker='X', s=200)
plt.title("K-Means Clustering Example")
plt.show()
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

Answer:
Inheritance allows a class (child) to reuse attributes and methods from another class (parent).



```python

class Animal:
    def speak(self):
        print("This animal makes a sound.")
class Dog(Animal):
    def speak(self):
        print("The dog barks.")
a = Animal()
a.speak()

d = Dog()
d.speak()
```

Question 10: How can time series analysis be used for anomaly detection?

Answer:
Time series analysis detects anomalies by comparing observed values against expected temporal patterns.
- Techniques include:
- Forecasting models (ARIMA, LSTM): Predict future values and flag deviations.
- Seasonal decomposition: Identify anomalies by separating trend and seasonality.
- Statistical thresholds: Detect sudden spikes or drops beyond confidence intervals.
- Applications: Fraud detection in transactions, monitoring server performance, and identifying unusual stock price movements.
